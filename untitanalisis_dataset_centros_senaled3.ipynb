```python
df_sin_duplicados.columns
```

```python
numero_centros_por_departamento = df_sin_duplicados.groupby('NOMBRE_DEPARTAMENTO')['CENTRO_FORMACION'].count()
numero_centros_ordenados = numero_centros_por_departamento.sort_values(ascending=False)
print(numero_centros_ordenados)
```

```python
plt.figure(figsize=(10, 8))
numero_centros_ordenados.plot(kind='bar', color='skyblue')
plt.title('Número de Centros de Aprendizaje por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Número de Centros')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**Distribucion por Departamento**



 La gráfica de barras muestra la cantidad de centros de aprendizaje por departamento. Se observa una mayor concentración en departamentos como Antioquia y Bogotá D.C., lo que sugiere una tendencia a concentrar más centros en áreas más pobladas o de mayor actividad económica.
